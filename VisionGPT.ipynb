```python
import streamlit as st
import sounddevice as sd
import numpy as np

def main():
    def record_audio(duration):
        fs = 44100  # sample rate
        recording = sd.rec(int(fs * duration), samplerate=fs, channels=2, dtype=np.int16)
        sd.wait()
    
        return recording
    
    st.title("Streamlit Audio Recorder")
    
    duration = st.slider("Select recording duration (seconds):", min_value=1, max_value=10, value=5)
    
    if st.button("Record"):
        st.write("Recording...")
        recording = record_audio(duration)
        st.write("Recording complete!")
    
        # Do something with the recording, e.g., play it back
        st.audio(recording, format="audio/wav")


if __name__ == "__main__":
    main()
```

```python
import google.generativeai as genai
```

```python
import os
from dotenv import load_dotenv

load_dotenv() 

genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
def get_gemini_response(prompt, img,input):
    model=genai.GenerativeModel('gemini-pro-vision')
    response=model.generate_content([prompt, img,input])
    return response.text
```

```python
import requests

# TEXT_TO_AUDIO_API_URL = "https://api-inference.huggingface.co/models/facebook/mms-tts-eng"
TEXT_TO_AUDIO_API_URL = "https://api-inference.huggingface.co/models/espnet/kan-bayashi_ljspeech_vits"
HUGGING_FACE_API_KEY=os.getenv("HUGGING_FACE_API_KEY")
headers = {"Authorization": HUGGING_FACE_API_KEY}

def text_to_audio(payload):
	response = requests.post(TEXT_TO_AUDIO_API_URL, headers=headers, json=payload)
	return response.content
```

```python
import wave

def save_audio_to_wav(audio_bytes, file_path, sample_width=2, channels=1, frame_rate=44100):
    with wave.open(file_path, 'wb') as wave_file:
        wave_file.setnchannels(channels)
        wave_file.setsampwidth(sample_width)
        wave_file.setframerate(frame_rate)
        wave_file.writeframes(audio_bytes)

# Example usage:
audio_bytes = audio_bytes  # Replace this with your actual audio data
file_path = 'output.wav'
save_audio_to_wav(audio_bytes, file_path)
```

# Audio to Text

```python
import requests

AUDIO_TO_TEXT = "https://api-inference.huggingface.co/models/openai/whisper-tiny"

def audio_to_text(filename):
    with open(filename, "rb") as f:
        data = f.read()
    response = requests.post(AUDIO_TO_TEXT, headers=headers, data=data)
    return response.json()
```
